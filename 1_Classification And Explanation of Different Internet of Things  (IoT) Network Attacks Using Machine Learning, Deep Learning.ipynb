```python
#one hard encoding
from sklearn.preprocessing import OneHotEncoder
import numpy as np

size_mapping = {
    'normal':0,
    'scanning':1,
    'dos':2,
    'injection':3,
    'ddos':4,
    'password':5,
    'xss':6,
    'ransomware':7,
    'backdoor':8,
    'mitm':9,

}
dataset['type'] = dataset['type'].map(size_mapping)


# 創建 OneHotEncoder
onehotencoder = OneHotEncoder(dtype=np.integer)

# 適應和轉換資料
onehotencoder.fit(dataset[["type"]])
type_encoded = onehotencoder.transform(dataset[["type"]]).toarray()

# 輸出轉換後的結果
# type_encoded
dataset[['normal', 'scanning', 'dos', 'injection', 'ddos', 'password', 'xss', 'ransomware', 'backdoor', 'mitm']] = type_encoded
dataset
```

#發現哪些特徵與其他一些特徵具有高度相關性

```python
import pandas as pd

# 假設 dataset 是一個 Pandas DataFrame
df = pd.DataFrame(dataset)

# 計算特徵之間的相關係數矩陣
corr_matrix = df.corr()

# 設定相關係數閾值
threshold = 0.9

# 找到具有高度相關性的特徵
highly_correlated_features = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) >= threshold:
            feature_i = corr_matrix.columns[i]
            feature_j = corr_matrix.columns[j]
            highly_correlated_features.append((feature_i, feature_j))

highly_correlated_features.remove(('label', 'normal'))
# 顯示具有高度相關性的特徵對
for feature_pair in highly_correlated_features:
    print(feature_pair)

print(len(highly_correlated_features))
```

#刪除高度相關性的特徵值

```python
# 從資料集中排除具有高度相關性的特徵
for feature_pair in highly_correlated_features:
    feature_i, feature_j = feature_pair
    if feature_i in df.columns and feature_j in df.columns:
        print(feature_j)
        df.drop(feature_j, axis=1, inplace=True)

# 將更新後的資料集存回原本的 dataset 變數中
dataset = df
# 更新後的資料集
print(dataset)
```

#KNN

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the range of k values to test
k_values = range(5, 13)

# Split the data into train and test sets
X = dataset.drop(['normal', 'scanning', 'dos', 'injection', 'ddos', 'password', 'xss', 'ransomware', 'backdoor', 'mitm', 'type', 'label'], axis=1)
y = dataset[['normal', 'scanning', 'dos', 'injection', 'ddos', 'password', 'xss', 'ransomware', 'backdoor', 'mitm']].astype(int)
# x,y = make_moons(noise=0.3)
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
# Initialize lists to store accuracy values
accuracies = []

# Iterate over each k value
for k in k_values:
    # Create and train the KNN classifier
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)

    # Predict the labels for the test set
    y_pred = model.predict(x_test)

    # Calculate and store the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the accuracy values
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy of KNN for different k values no normalization')
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical

X = dataset.drop(['normal', 'scanning', 'dos', 'injection', 'ddos', 'password', 'xss', 'ransomware', 'backdoor', 'mitm', 'type', 'label'], axis=1)
y = dataset[['normal', 'scanning', 'dos', 'injection', 'ddos', 'password', 'xss', 'ransomware', 'backdoor', 'mitm']].astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

category = len(dataset['type'].unique())  # Number of classes
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=40, activation='relu'))
model.add(tf.keras.layers.Dense(units=40, activation='relu'))
model.add(tf.keras.layers.Dense(units=50, activation='relu'))
model.add(tf.keras.layers.Dense(units=category, activation='softmax'))

model.compile(loss=tf.keras.losses.categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    batch_size=64,
                    epochs=50,
                    verbose=1)
```

```python
# 測試
score = model.evaluate(X_test, y_test, batch_size=128)
# 輸出結果
print("score:",score)

# 預測
predict = model.predict(X_test)
print("Ans:",np.argmax(predict[0]),np.argmax(predict[1]),np.argmax(predict[2]),np.argmax(predict[3]))

print("y_test",y_test[:4])

#保存模型架構
#with open("model.json", "w") as json_file:
#   json_file.write(model.to_json())
#保存模型權重
model.save("model-cifar")
```
